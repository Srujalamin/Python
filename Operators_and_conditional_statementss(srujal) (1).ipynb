```python
user_input = input('enter your age: ')

if user_input.isnumeric():
  user_input = int(user_input)

  if user_input == 14:
    print('you are not eligible. please try after 4 years.')
  elif user_input == 15:
    print('you are not eligible. please try after 3 years.')
  elif user_input == 16:
    print('you are not eligible. please try after 2 years.')
  elif user_input == 17:
    print('you are not eligible. please try after 1 years.')
  elif user_input == 18:
    print('you are eligible to apply for driving license.')
  elif user_input >= 60:
    print('you are not eligible to due to max age limit.')
  else:
    print('not in range.')
else:
  print('invalid input. please enter numeric value only.')
```

```python
user_input = input('enter your age: ')
req_min_age = 18
req_max_age = 60

if user_input.isnumeric():
  user_input = int(user_input)

  diff_min_age = req_min_age - user_input
  diff_max_age = user_input - req_max_age

  if user_input >= req_min_age and user_input <= req_max_age:
    print('you are eligible to apply for driving license.')
  elif user_input < req_min_age:
    print(f'you are not eligible. Please try after {diff_min_age} year(s).')
  elif user_input > req_max_age:
    print(f'you are not eligible. Your max age should less or equal {req_max_age} year(s).')
else:
  print('invalid input. please enter numeric value only.')
```
