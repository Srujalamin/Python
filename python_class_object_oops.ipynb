```python
class Person:
  name = None

  def set_info(self, name, city, country):
    self.name = name
    self.city = city
    self.country = country
  
  def get_info(self):
    print(self.name)
    print(self.city)
    print(self.country)
```

```python
p = Person
```

```python
p.get_info()
```

```python
class Person:
  def __init__(self, name, city, country):
    print('init called')
    self.name = name
    self.city = city
    self.country = country
  
  def get_info(self):
    print(self.name)
    print(self.city)
    print(self.country)
```

```python
p = Person('srujal' , 'ahmedabad' , 'india')
```
