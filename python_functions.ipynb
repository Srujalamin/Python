```python
info(name='srujal', city="ahmedabad").values()
```

```python
def info(name, city, **other):
  data = {'name': name, 'city': city}
  for i in other:
    data.setdefault(i, other[i])
    
  return data
```

```python
info(name='srujal', city="ahmedabad")
```

```python
info(name='srujal', city="ahmedabad", country="india")
```

```python
def info(name, city, **other):
  data = {'name': name, 'city': city}
  data.update(other)
  return data
```

```python
info(name='srujal', city="ahmedabad", country="india")
```
